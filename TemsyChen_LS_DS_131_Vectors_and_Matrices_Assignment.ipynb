## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
#Find ||g|| manually typing out the equation

g = [1,1,1,8]
h = [3,3,3,3]

norm_g = ((g[0]**2)+(g[1]**2)+(g[2]**2)+(g[3]**2))**.5
print("||g|| =", norm_g)

norm_h = ((h[0]**2)+(h[1]**2)+(h[2]**2)+(h[3]**2))**.5
print("||h|| =", norm_h)
```

```
#Now finding it with the numpy function, L2norm

print("||g|| =", np.linalg.norm(g))
print("||h|| =", np.linalg.norm(h))

#Got the same answer as above
```

```
#Now finding it with the numpy function, L1norm

np.linalg.norm(g, ord=1)

#With Vignesh's help: so it turns out np.linalg.norm uses ord=2 as the default. You must specify ord=1 for the L1norm.
#Also I have to pass g, 
```

C = \begin{bmatrix}

\qquad
D = \begin{bmatrix}

0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
\qquad
E = \begin{bmatrix}
1 & 3 \\
5 & 7
\end{bmatrix}
\end{align}

**Dimensions are: A=3x2, B=1x3, C=2x3, D=3x3, E=2x2**

**Possible multiplication combinations: BA, AC, BD, DB, CD, AE, EC**

```
#Create the matrices

A = np.array([[1,2],
              [3,4],
              [5,6]])

B = np.array([2,4,6])

C = np.array([[9,6,3],
              [4,7,11]])

D = np.array([[1,0,0],
              [0,1,0],
              [0,0,1]])

E = np.array([[1,3],
              [5,7]])
```

## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```
print("CD =\n", np.matmul(C,D))
print("C dimensions = 2x3, D dimensions = 3x3. CD dimensions = 2x3\n")

print("AE =\n", np.matmul(A,E))
print("A dimensions = 3x2, E dimensions = 2x2. CD dimensions = 3x2\n")

print("BA =\n", np.matmul(B,A))
print("B dimensions = 1x3, A dimensions = 3x2. CD dimensions = 1x2")
```

**The resulting matrix from multiplying matrices is the dimension of the row of the first matrix, and the dimension of the column on the second matrix**

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
H=np.array([[12,11],[7,10]])

J=np.array([[0,1,2],[7,10,4],[3,2,0]])
```

```
#Find determinant of H

det_H = 12*10-11*7
print("|H| =", det_H)

det_J = (0*10*0)+(1*4*3)+(2*7*2)-(2*10*3)-(1*7*0)-(0*4*2)
print("|J| =", det_J)
```

```
#Check using function

print(np.linalg.det(H))
print(np.linalg.det(J))

#Different, but very close. That's weird.
```
